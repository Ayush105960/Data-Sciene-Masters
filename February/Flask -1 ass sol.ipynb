    Flask is a micro web framework written in Python. It is classified as a microframework because it does not require particular tools or libraries. It has no database abstraction layer, form validation, or any other components where pre-existing third-party libraries provide common functions

Advantages of Flask Framework : 
    
        * Light weight and easy to use.
        * Scalable
        * Flexible

    App routing is a mapping method for accessing the functions within server using url in any device .    App routing is the method for mapping urls to specific functions that will handle the logic of  the url 

Reasons to use app routes :     
    
    App routes are used so that the specific functions ,which is being routed , can be accessible from any system irrespective of the programming language . 
    App routing is used to map the specific URL with the associated function that is intended to perform some task. It is used to access some particular page .
    App routes basically makes the function accessible through url instead of the respective calling methods in respective programming langs.

## 5. 

route() function is used for url binding in flask .

url_for() method is used for dynamically allocating urls to the view functions.

```python
from flask import *  
   
app = Flask(__name__)  
  
@app.route('/admin')  
def admin():  
    return 'admin'  
  
@app.route('/librarion')  
def librarion():  
    return 'librarion'  
  
@app.route('/student')  
def student():  
    return 'student'  
  
@app.route('/user/<name>')  
def user(name):  
    if name == 'admin':  
        return redirect(url_for('admin'))  
    if name == 'librarion':  
        return redirect(url_for('librarion'))  
    if name == 'student':  
        return redirect(url_for('student'))  
if __name__ =='__main__':  
    app.run(debug = True)  
```
