POST:- 
    
        It is also an http method to send data from client to server but without using url . It sends data or arguement in a secure way i,e. through a packet.This method makes enables users to send data over to the server. e.g In any authentication-enabled application, the registration and login form is the best example for the post method. Whenever we enter Information and submit the data get transferred over to the POST request. And if we noticed, unlike get request there is no information will be ever displayed in the URL.

## 3.

redirect() method is basically used in flask to redirect the path to another route.<br>
Syntax: 

    from flask import redirect 
    
    redirect(<location/url> , <status code>, <response>

Flask redirect is defined as a function or utility in Flask which allows developers to redirect users to a specified URL and assign a specified status code. When this function is called, a response object is returned, and the redirection happens to the target location with the status code. When a website is visited, server receives a request and after processing responds to the request with a three-digit code known as the HTTP code. In a programming world this status code that is returned as a response is known as Errors. Errors and redirect function goes hand in hand.

```python
from flask import Flask , request , render_template , redirect , abort ,url_for

app = Flask(__name__)

@app.route("/home/<user>")
def home(user):
    if user == "admin" :
        return redirect(url_for("AdminLoginPage") )  # url_for detects the function within the server not the url
    else:
        abort(401)
        
@app.route("/AdminLoginPage")
def AdminLoginPage():
    return "redirected to admion login page"
if __name__ == "__main__":
    app.run(host = "0.0.0.0" )
```

templates in flask are the html pages which will be used for front end side.

    Templates are files that contain static data as well as placeholders for dynamic data. A template is rendered with specific data to produce a final document. Flask uses the Jinja template library to render templates.

render_template() is used for rendering the specified file to the client with the static data as well as dynamic data placed at the placeholder.

    Flask provides us the render_template() function which can be used to render the external HTML file to be returned as the response from the view function.

```python
from flask import Flask , request , render_template , redirect , abort ,url_for

app = Flask(__name__)

@app.route("/home/<user>")
def home(user):
    if user == "admin" :
        return redirect(url_for("AdminLoginPage") )
    else:
        abort(401)
        
@app.route("/LoginPage")
def AdminLoginPage():
    return "<h1 style = 'color: red ;text-align : center;'> redirected to Admin login page </h1>"
if __name__ == "__main__":
    app.run(host = "0.0.0.0" , port=8080 )
```
