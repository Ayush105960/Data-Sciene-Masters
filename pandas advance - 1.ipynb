```python
df = pd.DataFrame({'Text' : [['one'] , ['two', 'three'] , [322,232,534,5354]]})
```

```python
df
```

```python
def no_words(df):
    index = 0
    df['Word_Count'] = None
    for each in df['Text'].values:
        df['Word_Count'][index] = len(each)
        index += 1
    return df
```

```python
df = no_words(df)
```

```python
df = pd.DataFrame({'Email' : ['abc@gmail.com' , 'ayush@gmail.com' , 'bec@gmail.com']})
```

```python
def username(df):
    df['Username'] = None
    index = 0 
    for each in df['Email']:
        df['Username'][index] = each.split('@')[0]
        index += 1
    return df
```

```python
df = username(df)
```

```python
df
```

```python
import numpy as np 
import datetime as dt
```

```python
df = pd.DataFrame({'Sales' : np.random.random(10) * 1000000 , 'Date' : pd.date_range('2023-1-1' , periods = 10 , freq="3D" , tz = "Asia/Tokyo")})
```

```python
df
```

```python
def average_calc(df):
    df['MovingAverage'] = None
    index = 0
    for each in df['Date']:
        past7 = each - dt.timedelta(days = 7)
        df_filtered = df[(df['Date'] >= past7) & (df['Date'] <= each)]
        df['MovingAverage'][index] = df_filtered['Sales'].mean()
        index += 1
    return df
```

```python
df = average_calc(df)
df
```

## 11. 

```python
df = pd.DataFrame()
df['date'] = pd.date_range('2023-1-1', '2023-3-1', freq='3D' )
print(df)
```

```python
def day_teller(df):
    df["WeekDay"] = None
    index = 0
    for each in df['date']:
        df["WeekDay"][index] = each.day_name()
        index += 1
    return df
```

```python
df = day_teller(df)
```
